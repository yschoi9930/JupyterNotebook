2. 상품배치와 상품 구매 금액에 따른 관계 분석
    - 상품 배치에 따른 상품 구매 금액 평균 차이 분석(상품 구매 금액 0원 미포함)
    - 상품 배치에 따른 상품 구매 금액 평균 차이 분석(상품 구매 금액 0원 포함)
    - 구매여부와 상품 배치 간 독립성 파악

```python
placement_A = pd.read_csv('../Data/AB테스트/상품배치_A.csv', engine='python', index_col='고객ID')
placement_A.head()
placement_A.tail()
```

```python
placement_B = pd.read_csv('../Data/AB테스트/상품배치_B.csv', engine='python', index_col='고객ID')
placement_B.head()
placement_B.tail()
```

```python
placement_C = pd.read_csv('../Data/AB테스트/상품배치_C.csv', engine='python', index_col='고객ID')
placement_C.head()
placement_C.tail()
```

```python
# 구매 금액이 0인 데이터는 제외 시키고 작업
```

```python
A_out_zero = placement_A.loc[placement_A['구매금액'] !=0]['구매금액'].values
B_out_zero = placement_B.loc[placement_B['구매금액'] !=0]['구매금액'].values
C_out_zero = placement_C.loc[placement_C['구매금액'] !=0]['구매금액'].values
```

```python
# 세 그룹간의 평균 차이 확인(ANOVA 검정 진행)
from scipy.stats import *
print(kstest(A_out_zero,'norm'))
print(kstest(B_out_zero,'norm'))
print(kstest(B_out_zero,'norm'))
# pvalue : 0.0 정규분포다
```

```python
# 일원분산분석
print(f_oneway(A_out_zero,B_out_zero,C_out_zero))
# 세 그룹의 평균은 같지 않다
```

```python
# 사후분석
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
group = ['A']*len(A_out_zero) + ['B']*len(B_out_zero) + ['C']*len(C_out_zero)
data = A_out_zero.tolist() + B_out_zero.tolist() + C_out_zero.tolist()
print(pairwise_tukeyhsd(data,group))
# B가 다름을 보여줌
```

```python
# 구매금액 0을 포함하여 같은 작업을 수행
```

```python
# 구매 여부와 상품배치 간 관계 파악
```

```python
# 데이터 변환
placement_A['상품배치'] = 'A'
placement_B['상품배치'] = 'B'
placement_C['상품배치'] = 'C'
```

```python
placement = pd.concat([placement_A,placement_B,placement_C],axis=0,ignore_index=False)
placement
```

```python
placement['구매여부'] = (placement['구매금액']!=0).astype(int)
placement.head()
```

```python
# 교차테이블 생성
cross_tab = pd.crosstab(placement['상품배치'],placement['구매여부'])
cross_tab
```

```python
obs=cross_tab.values # array 형태로
statistics,pvalue,dof,expected = chi2_contingency(obs,correction=False)
print(pvalue) # 0.068 귀무가설 채택 / 구매여부와 상품배치에는 관계가 있다고 보기 힘듦
```

3. 사이트맵 구성에 따른 체류 시간 차이 분석
    - 사이트맵 별 체류시간 평균 계산
    - 사이트맵에 따른 체류 시간 평균 차이 분석
    - 평균차이 분석 시각화

```python
sitemap_A = pd.read_csv('../Data/AB테스트/사이트맵_A.csv', engine='python')
```

```python
sitemap_B = pd.read_csv('../Data/AB테스트/사이트맵_B.csv', engine='python')
```

```python
sitemap_C = pd.read_csv('../Data/AB테스트/사이트맵_C.csv', engine='python')
```

```python
A_time = sitemap_A['체류시간(분)'].values
A_time
```

```python
B_time = sitemap_B['체류시간(분)'].values
A_time
```

```python
C_time = sitemap_C['체류시간(분)'].values
C_time
```

```python
# 일원분산분석
from scipy.stats import *
print(kstest(A_time,'norm'))
print(kstest(B_time,'norm'))
print(kstest(C_time,'norm'))
```

```python
print(f_oneway(A_time,B_time,C_time)) # 평균 차이는 존재하지 않는다
```
