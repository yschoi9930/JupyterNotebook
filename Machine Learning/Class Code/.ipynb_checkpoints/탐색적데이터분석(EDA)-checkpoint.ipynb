### 탐색적분석(EDA)
- 데이터의 전체적인 특성을 살펴보는 것
- 수집한 데이터가 적절한지 알아보는 과정
- 데이터가 통계적인 가정을 만족하고 있는지를 확인하고 있는지 확인
    - 분포가 골고루 있는지(남여 숫자가 비슷한지, 사는지역 계층등이 비슷하게 데이터가 수집되어 있는지)
    - 최종분석하는 상황이나 모델에 쓰는 데이터가 비율이 맞게 들어 있는 지 등을 확인
- 데이터의 양적 측면 뿐 아니라 질적 측면도 확인하는 과정

- 데이터 전처리를 같이 수행
    - 데이터 클리닝 - 결측치 처리, 틀린값 처리
    - 스케일링 - 여러 변수 값의 범위를 동등한 조건으로 맞추는 것: 표준 스케일링, min-max 스케일링        
    - 이상치 처리 - 이상치 탐지 및 제거
    - 데이터 변환 - 로그 변환, 카테고리 인코딩 등

### 목적변수

```python
 # 박스플롯
plt.boxplot(df['SalePrice'])    ## 목적변수에 튀는값이 상당히 있음
```

```python
# 기초 통계값
df['SalePrice'].describe()
```

```python
# 히스토그램 : 발생 빈도수
plt.hist(df['SalePrice'],bins=30)
plt.show()
```

```python
## 확률분포보기 : 가장 가까운 정규분포와 비교
## 정규분포 얼마나 가까운지 확인
sns.distplot(df['SalePrice'],fit=norm)
```

```python
### 정규분포와 가까우면
    - 데이터가 정규분포를 따르면 평균값, 분산 예측하는 것이 수월하다
    - skewness(), kurtosis() 이 값들이 0에 가까울수록 정규분포와 유사하다
```

```python
print("왜도 : ", df['SalePrice'].skew())
print("첨도 : ", df['SalePrice'].kurtosis())
```

```python
##  정규분포의 왜도와 첨도 
N=1000
x=np.random.randn(N)
sns.distplot(x, fit=norm)
print("왜도 : ", stats.skew(x))
print("첨도 : ", stats.kurtosis(x))
```

```python
# 목적변수 돈과 관련 있으면 log 변환 - 정규 분포와 꽤 비슷해 짐
sns.distplot(np.log(df['SalePrice']),fit=norm)
```

```python
print("왜도 : ", np.log(df['SalePrice']).skew())
print("첨도 : ", np.log(df['SalePrice']).kurtosis())
```

```python
###  목적변수와 입력특성들의 상관계수
- 데이터 분석은 입력변수를 사용해서 목적변수를 예측하는 작업
- 입력변수와 목적변수 사이의 상관계수를 보면 연관성을 파악하는데 도움이 된다
    - 중고차가격 : 연식이 오래되면 가격이 떨어진다
- 입력변수 - 특성(features), 설명변수, 독립변수
- 목적변수 - 레이블, 출력변수, 종속변수
```

```python
## 히트맵을 이용한 상관 계수 확인
corrmat = df.corr() # 모든 변수들을  쌍대 쌍으로 상관계수를 구해 매트릭스로 표현
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(corrmat)
## 목적변수와 상관계수가 큰 특성을 찾는다
```

## 이상치 찾기
- 표준 스케일링 후 편차가 큰 데이터를 찾는다
- log 취한 값을 표준 스케일링 한 후 이상치 데이터를 찾는다(범위를 -3 보다 작은 +3 보다 큰)
- 산포도를 그린 후 너무 멀리 떨어진 데이터를 찾기

```python
## 표준 스케일링
df2 ["Sale_scaled"]=StandardScaler().fit_transform(df2[['SalePrice']])
```
