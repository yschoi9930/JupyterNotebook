```python
# 6. 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다.
a=(15,1,"a",100)
b=(16,2,"b",90)
c=(15,3,"c",80)
d=(14,4,"d",70)
df = pd.DataFrame([a,b,c,d],columns=["나이","번호","이름","점수"])
df
```

```python
# 6-1.학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, 
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
data = {'나이' : [15,16,15,14],
        '번호' : [1,2,3,4],
        "이름" : ['a','b','c','d'],
        '점수' : [100,90,80,70]}
df=pd.DataFrame(data)
df

index = ['a','b','c','d']
columns=["나이","번호","점수"]
df_score1 = pd.DataFrame(data, index=index, columns = columns)
df_score1

#-----------------------------------------------나의 풀이

# 6-1.학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, 
df_score2 = df[["나이","번호","점수"]]
df_score2
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
df_score2.index = df['이름']
df_score2

# reset_index 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.
df_score3 = df_score2.reset_index()
df_score3
```

```python
# 6-2.학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 set_index 명령을 
# 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
df_score3
df_score3 = df_score3.set_index('이름')
df_score3
```

```python
# 7. A 반 학생 5명과 B반 학생 5명의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다.
# "반", "번호", "국어", "영어", "수학" 을 열로 가지는 데이터프레임 df_score3을 만든다.
# df3
# 번호	반	국어	영어	수학			
# ID0	A	24	21	90
# ID1	A	3	38	22
# ID2	A	56	96	66
# ID3	A	72	20	2
# ID4	A	0	44	63
# ID0	B	21	93	60
# ID1	B	19	39	1
# ID2	B	74	14	51
# ID3	B	41	26	90
# ID4	B	10	81	69

np.random.seed(3)
df3 = pd.DataFrame({
    '반' : ['A'] * 5 +['B'] * 5,
    '번호' : ['ID' + str(i) for i in range(5)] *2,
    '국어' : np.random.randint(0,101,10),
    '영어' : np.random.randint(0,101,10),
    '수학' : np.random.randint(0,101,10),})
df3

#(0,101,10) - > 0에서 100까지 사이의 숫자를 10개(size)
```

```python
# 7-1 데이터 프레임 df_score3에  df3을 사용하여 각 학생의 평균을 나타내는 행을 오른쪽에 추가한 후 저장한다

df3['평균'] = ((df3['국어'] + df3['영어'] + df3['수학'])/3).round(2)
df3

df3['평균'] = df3.mean(axis=1).round(2)
df3
```

```python
# 7-2 데이터 프레임 df_score4에  df3을 사용하여 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.

df_score3 = df3
df_score3.loc['평균'] = df_score3.mean().round(2)
df_score3
```
